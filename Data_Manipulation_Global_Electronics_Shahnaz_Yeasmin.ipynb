### Identify Data Type Issues

**Q1.4** - Identify the potential data type issues. Give the names of the columns which have these issues. **(2 marks)**

```python
#Write your code here
def detect_data_issues(df):
    issues = {}
    for col in df.columns:
        # Check for object columns that should be numeric
        if df[col].dtype == 'object':
            # Try converting to numeric and see how many fail
            non_numeric = pd.to_numeric(df[col], errors='coerce').isna().sum()
            if non_numeric > 0:
                issues[col] = f"{non_numeric} non-numeric entries"
                
        # Check for missing values
        if df[col].isna().sum() > 0:
            if col in issues:
                issues[col] += f", {df[col].isna().sum()} missing values"
            else:
                issues[col] = f"{df[col].isna().sum()} missing values"
    return issues

```

###	Find Revenue Growth Metric

* **Q3.2.1** - Which metric can be used to measure the company’s revenue growth over month on month (come up with a metric)? **(1 mark)**
* **Q3.2.2 and 3.2.3** - How can we calculate this metric for each store (write a Python code)?  **(3 marks)**
* **Q3.2.4** - What insights and recommendations can be derived from this metric?  **(2 marks)**

**Note** - Give atleast 1 insight and 1 recommendation



'Order Date' can be utilised to measure revenue growth MoM


```python
#Write your code here
#Q3.2.1
# Extract year-month
sales_full['YearMonth'] = sales_full['Order Date'].dt.to_period('M')

# Aggregate revenue by month
monthly_revenue = sales_full.groupby('YearMonth')['Revenue_USD'].sum().sort_index()

# Calculate Month-on-Month growth
monthly_growth = monthly_revenue.pct_change() * 100  # percentage change
monthly_growth = monthly_growth.round(2)  # round to 2 decimals

# Combine revenue and growth in one table
monthly_summary = pd.DataFrame({
    'Revenue_USD': monthly_revenue,
    'MoM_Growth_%': monthly_growth
})

print(monthly_summary)
```

```python

# Find the month(s) with the highest revenue
max_revenue = monthly_summary['Revenue_USD'].max()
top_months = monthly_summary[monthly_summary['Revenue_USD'] == max_revenue]

print("Month(s) with highest revenue:")
print(top_months)
```

### Perform Store Age–Based Segmentation

* **Q3.3.1 and 3.3.2** - Calculate the age of the stores in 5 buckets ('0–5 years', '5–10 years', '10–15 years', '15–20 years', '20+ years'). **(3 marks)**
* **Q3.3.3** - Calculate the total revenue in each bucket (write a Python code). **(2 marks)**

**Note** - Give atleast 1 insight and 1 recommendation


```python
#Write your code here 

#Q3.3.1 and 3.3.2 

from datetime import datetime            

# Calculate age in years assuming current year = 2021
current_year = 2021

sales_full['Store_Age'] = current_year - sales_full['Open Date'].dt.year

# Define bins and labels
bins = [0, 5, 10, 15, 20, float('inf')]
labels = ['0–5 years', '5–10 years', '10–15 years', '15–20 years', '20+ years']

# Categorize into buckets
sales_full['Store_Age_Bucket'] = pd.cut(sales_full['Store_Age'], bins=bins, labels=labels, right=False)

sales_full[['Open Date', 'Store_Age', 'Store_Age_Bucket']].head()
```

```python
#Write your code here        
#Q3.3.3

revenue_by_bucket = sales_full.groupby('Store_Age_Bucket')['Revenue_USD'].sum().sort_index()
print(revenue_by_bucket)
```

```python
#Q3.3.4
avg_revenue_per_store = store_revenue.groupby('Store_Age_Bucket')['Revenue_USD'].mean()
print(avg_revenue_per_store)
```
